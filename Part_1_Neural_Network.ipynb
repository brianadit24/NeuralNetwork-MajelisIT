# Note on Weights
Neural Network memiliki weight yang sangat-sangat banyak. Berbeda dengan linear regression yang hanya mencari kombinasi $a$ & $b$ yang menghasilkan loss sekecil mungkin, Neural network bisa memiliki weight dari ratusan hingga jutaan sehingga diperlukan data yang lebih banyak pula.
